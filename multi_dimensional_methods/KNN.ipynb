Pre-processing and Feature Engineering

Отмасштабируем данные и применим анализ основных компонентов (PCA), чтобы выделить наиболее важные функции для дальнейшего использования в обучающих моделях. </br>
Обработка данных такого размера требует значительных вычислительных ресурсов (219521, 53), отсюда и причина уменьшения размерности с помощью PCA.

```python
# Стандартизируем/масштабируем набор данных и примените PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline

df2 = df.drop(['machine_status'], axis=1)
names=df2.columns
x = df[names]

scaler = StandardScaler()
pca = PCA()
pipeline = make_pipeline(scaler, pca)
pipeline.fit(x)
```

```python
features = range(pca.n_components_)
_ = plt.figure(figsize=(22, 5))
_ = plt.bar(features, pca.explained_variance_)
_ = plt.xlabel('PCA feature')
_ = plt.ylabel('Variance')
_ = plt.xticks(features)
_ = plt.title("Важность основных компонентов")
plt.show()
```

```python
# Расчет PCA с 2 компонентами
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])
```

```python
df['pc1']=pd.Series(principalDf['pc1'].values, index=df.index)
df['pc2']=pd.Series(principalDf['pc2'].values, index=df.index)
```

#### Кластеризация K-средних

Стратегия:

Основное предположение в обнаружении аномалий на основе кластеризации состоит в том, что если мы кластеризуем данные, нормальные данные будут принадлежать кластерам, в то время как аномалии не будут принадлежать ни одному кластеру или принадлежать малым кластерам. 

- Вычислим расстояние между каждой точкой и ближайшим к ней центроидом. Самые большие расстояния считаются аномальными.
- Используем outliers_fraction, чтобы предоставить алгоритму информацию о доле выбросов, присутствующих в нашем наборе данных. Ситуации могут варьироваться от набора данных к набору данных. Однако в качестве исходной цифры outliers_fraction=0,14 (14% df являются выбросами, как показано выше).
- Вычислим number_of_outliers с помощью outliers_fraction.
- Установим порог как минимальное расстояние этих выбросов.
- Аномальный результат аномалии1 содержит указанный выше метод Cluster (0: нормальный, 1: аномальный).
- Визуализируем аномалии с помощью кластерного представления.
- Визуализируем аномалии с помощью представления временных рядов.

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(principalDf.values)
labels = kmeans.predict(principalDf.values)
unique_elements, counts_elements = np.unique(labels, return_counts=True)
clusters = np.asarray((unique_elements, counts_elements))
```

```python
_ = plt.figure(figsize = (9, 7))
_ = plt.bar(clusters[0], clusters[1], tick_label=clusters[0])
_ = plt.xlabel('Кластеры')
_ = plt.ylabel('Количество точек')
_ = plt.title('Количество точек в кажом кластере')
plt.show()
```

```python
_ = plt.figure(figsize=(9,7))
_ = plt.scatter(principalDf['pc1'], principalDf['pc2'], c=labels)
_ = plt.xlabel('pc1')
_ = plt.ylabel('pc2')
_ = plt.title('K-means кластеризации')
plt.show()
```

```python
def getDistanceByPoint(data, model):
    """ Функция, вычисляющая расстояние между точкой и центроидом кластера,
             возвращает расстояния в pandas series"""
    distance = []
    for i in range(0,len(data)):
        Xa = np.array(data.loc[i])
        Xb = model.cluster_centers_[model.labels_[i]-1]
        distance.append(np.linalg.norm(Xa-Xb))
    return pd.Series(distance, index=data.index)
```

```python
# Предположим, что 13% всего набора данных являются аномалиями
outliers_fraction = 0.13
# получим расстояние между каждой точкой и ближайшим к ней центроидом. Самые большие расстояния считаются аномалиями
distance = getDistanceByPoint(principalDf, kmeans)
# number of observations that equate to the 13% of the entire data set
# количество наблюдений, которые соответствуют 13% всего набора данных
number_of_outliers = int(outliers_fraction*len(distance))
# В качестве порога берем минимум из самых больших 13% расстояний
threshold = distance.nlargest(number_of_outliers).min()
# anomaly1 содержит результат аномалии вышеописанного метода Cluster (0: нормальный, 1: аномальный)
principalDf['anomaly1'] = (distance >= threshold).astype(int)
```
